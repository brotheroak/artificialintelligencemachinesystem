```
import cv2
import numpy as np
import matplotlib.pyplot as plt
import time

image=cv2.imread("./sample_data/sample_image3.jpg")
hsv=cv2.cvtColor(image,cv2.COLOR_BGR2HSV)
lowerbound=np.array([30,0,0])
upperbound=np.array([60,255,255])
new=cv2.inRange(hsv,lowerbound,upperbound)
cv2.imwrite("./sample_data/test_hsv3.jpg",hsv)
cv2.imwrite("./sample_data/get_green.jpg",new)
plt.grid(False)
plt.imshow(new)
```
